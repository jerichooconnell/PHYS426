The orbitals in the surface wave do not move the wave forward as the wave is simply a propagation of energy through a medium that maintains energy and has a restoring force. The circular motion of the particles is just a byproduct of the pressure gradient brought about by the deformation of the surface. Using this logic a negative wave would be orbitals in the oposite direction of rotation.

## Q2

```python
import numpy as np
import matplotlib.pyplot as plt
%pylab

xlong = np.linspace(0,100,1001)
ylong =  np.sqrt(9.81*xlong* np.tanh(xlong*0.01))
ylong2 =  np.sqrt(9.81*xlong* np.tanh(xlong*0.001))
ylong3 =  np.sqrt(9.81*xlong* np.tanh(xlong*0.1))
ylong5 =  np.sqrt(9.81*xlong* np.tanh(xlong*0.00001))
ylong4 =  np.sqrt(9.81*xlong* np.tanh(xlong*1000.))
ylong4 =  np.sqrt(9.81*xlong* np.tanh(xlong*100000.))

plt.figure()
plt.plot(xlong,ylong)
plt.plot(xlong,ylong2)
plt.plot(xlong,ylong3)
plt.plot(xlong,ylong5)
plt.plot(xlong,ylong4)
plt.ylabel('w')
plt.xlabel('k')
```

As H gets very large and very small there are two limiting behaviors, which correspond to tanhx going to x for small numbers and to 1 for very large numbers. Thus the waves reach a max speed in the medium at large depths and go to very low speed for small depths.
